```python
class Student:
    def __init__(self, name, score):
        self.name = name
        self.score = score

    def __repr__(self):
        return f"Student(name={self.name}, score={self.score})"

    def add_score(self, score):
        self.score.append(score)

    def average_score(self):
        return sum(self.score) / len(self.score)

    def get_info(self):
        return f"Name: {self.name}, Scores: {self.score}, Average: {self.average_score():.2f}"

students = [
        Student("Alice", [65]),
        Student("Bob", [45]),
        Student("Charlie", [85]),
        ]

while True:
    print("\nกรอกตัวเลขเพื่อเลือกฟังก์ชัน")
    print("1. แสดงข้อมูลนักเรียน")
    print("2. เพิ่มคะแนนนักเรียน")
    print("3. คำนวณคะแนนเฉลี่ยของนักเรียน")
    print("4. ออกจากโปรแกรม")
    menu = int(input("เลือกฟังก์ชัน(กรอกเป็นตัวเลขเท่านั้น): "))

    if menu == 1:
            for student in students:
                print(student.get_info())
    elif menu == 2:
        name = input("กรอกชื่อนักเรียนที่ต้องการเพิ่มคะแนน: ")
        score = int(input("กรอกคะแนนที่เพิ่ม: "))
        for student in students:
            if student.name == name:
                student.add_score(score)
                print(f"เพิ่มคะแนน {score} ให้กับ {name} เรียบร้อยแล้ว")
                print("ข้อมูลล่าสุด:", student.get_info())
                break
        else:
            print("ไม่พบชื่อนักเรียนในระบบ หรือกรอกชื่อผิด")
    elif menu == 3:
            print("คะแนนเฉลี่ยของนักเรียนแต่ละคน:")
            for student in students:
                print(f"{student.name}: {student.average_score():.2f}")
    elif menu == 4:
        print("ออกจากโปรแกรม")
        break
    else:
        print("กรุณาเลือกเมนู 1-4 เท่านั้น")

class Classroom:
    def __init__(self, name, capacity):
        self.name = name
        self.capacity = capacity
        self.students = []

    def add_student(self, student):
        if len(self.students) < self.capacity:
            self.students.append(student)
            return True
        return False

    def insert_student(self, index, student):
        if len(self.students) < self.capacity and 0 <= index <= len(self.students):
            self.students.insert(index, student)
            return True
        return False

    def find_student(self, identifier):
        if isinstance(identifier, int):
            if 0 <= identifier < len(self.students):
                return self.students[identifier]
            return None
        elif isinstance(identifier, str):
            for student in self.students:
                if student.name == identifier:
                    return student
            return None
        return None

    def sort_students(self):
        self.students.sort(key=lambda s: s.name)

    def reverse_students(self):
        self.students.reverse()

    def remove_student(self, student_name):
        student_to_remove = self.find_student(student_name)
        if student_to_remove:
            self.students.remove(student_to_remove)
            return True
        return False

    def clear_students(self):
        self.students.clear()

    def __str__(self):
        return f"Classroom {self.name} with capacity {self.capacity}"

alice = Student("Alice", [])
alice.add_score(90)
alice.add_score(85)
alice.add_score(70)

bob = Student("Bob", [])
bob.add_score(60)
bob.add_score(75)
bob.add_score(80)

charlie = Student("Charlie", [])
charlie.add_score(100)
charlie.add_score(95)
charlie.add_score(90)

david = Student("David", [])
eva = Student("Eva", [])

math_class = Classroom("Math 101", 10)

math_class.add_student(alice)
math_class.add_student(bob)
math_class.add_student(charlie)
math_class.add_student(david)
math_class.add_student(eva)

print(f"{alice.name} - {alice.score} Average: {alice.average_score()}")
print(f"{bob.name} - {bob.score} Average: {bob.average_score()}")
print(f"{charlie.name} - {charlie.score} Average: {charlie.average_score()}")

print("All students:", [str(s) for s in math_class.students])

frank = Student("Frank", [])
math_class.insert_student(2, frank)
print("After insert:", [str(s) for s in math_class.students])

math_class.sort_students()
print("Sorted:", [str(s) for s in math_class.students])

math_class.reverse_students()
print("Reversed:", [str(s) for s in math_class.students])

try:
    bob_index = math_class.students.index(math_class.find_student("Bob"))
    print(f"Index of Bob: {bob_index}")
except ValueError:
    print("Bob not found.")

math_class.remove_student("Frank")
print("After remove:", [str(s) for s in math_class.students])

math_class.clear_students()
print("After clear:", math_class.students)
```
